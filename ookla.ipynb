Joining the two datasets based on "school_id_giga"

```python
merged_df = pd.merge(uzb_measuremts, uzb_school_geolocations, on='school_id_giga', how='inner')
```

```python
merged_df
```

```python
# Converting date time to the proper format for calculations purpose
merged_df['date'] = pd.to_datetime(merged_df['date'], format='%d.%m.%Y', errors='coerce')
```

```python
# We filter the data only for Q4 2024
merged_df = merged_df[(merged_df['date'] >= '2024-10-01') & (merged_df['date'] <= '2024-12-31')]
```

```python
merged_df
```

```python
#To find the unique schools
num_unique_values_school_id_giga = merged_df['school_id_giga'].nunique()
print(f"Number of unique values in 'school_id_giga': {num_unique_values_school_id_giga}")
```

```python
# Converting the columns to numeric, handling errors by setting non-convertible values to NaN
merged_df['download_speed'] = pd.to_numeric(merged_df['download_speed'], errors='coerce')
merged_df['upload_speed'] = pd.to_numeric(merged_df['upload_speed'], errors='coerce')
merged_df['latency'] = pd.to_numeric(merged_df['latency'], errors='coerce')

# Now perform the groupby and aggregation
averaged_data = merged_df.groupby('school_id_giga').agg({
    'download_speed': 'mean',
    'upload_speed': 'mean',
    'latency': 'mean'
})
```

```python
averaged_data
```

Adding latitude and longitude of the schools

```python
averaged_data = pd.merge(averaged_data, merged_df[['school_id_giga', 'latitude', 'longitude']], on='school_id_giga', how='left')
```

```python
averaged_data
```

After adding latitude and longitude, we see the number of rows are added and we need only one data point for one school for Q4 2024 as the Ookla datasets only has one data points for 3 months, we want to have one row for each unique school

```python
# Drop duplicates based on school ID before aggregation
unique_schools_df = merged_df.drop_duplicates(subset=['school_id_giga'])

# Now perform the groupby and aggregation on the unique schools DataFrame
averaged_data = unique_schools_df.groupby('school_id_giga').agg({
    'download_speed': 'mean',
    'upload_speed': 'mean',
    'latency': 'mean',
    'latitude': 'first',  # or 'last', depending on your preference
    'longitude': 'first'   # or 'last', depending on your preference
})
```

```python
# Project to a projected CRS (e.g., UTM) for accurate buffer calculations
csv_gdf = csv_gdf.to_crs(epsg=32642)  # Choose an appropriate UTM zone for your area
gdf = gdf.to_crs(epsg=32642) # Change gdf CRS too

# Create 3 km buffers
csv_gdf['buffer'] = csv_gdf.geometry.buffer(7000)  # 7000 meters = 7 km

# Create a GeoDataFrame from the buffers
buffer_gdf = gpd.GeoDataFrame(csv_gdf, geometry='buffer')
```

```python
# Perform a spatial join to find GPKG points within buffers
joined_gdf = gpd.sjoin(gdf, buffer_gdf, how='inner', predicate='intersects')

# Count the number of GPKG points within each buffer
count_within_buffers = joined_gdf.groupby(joined_gdf.index_right).size()

# Print the results
print(count_within_buffers)
```

Renaming columns for comparison

```python
# Rename specific columns
giga_schooldata = giga_schooldata.rename(columns={'avg_d_kbps': 'ookla_download_speed', 'avg_u_kbps': 'ookla_upload_speed','avg_lat_ms': 'ookla_lat_ms','download_speed': 'giga_school_download_speed','upload_speed': 'giga_school_upload_speed','latency': 'giga_school_latency'})
```

Converting ookla data to mbps for the sake of comparison

```python
giga_schooldata['ookla_download_speed'] = giga_schooldata['ookla_download_speed'] / 1000
giga_schooldata['ookla_upload_speed'] = giga_schooldata['ookla_upload_speed'] / 1000
giga_schooldata['ookla_lat_ms'] = giga_schooldata['ookla_lat_ms'] * 10
```

```python
giga_schooldata['download_speed_difference'] = giga_schooldata['giga_school_download_speed'] - giga_schooldata['ookla_download_speed']
giga_schooldata['upload_speed_difference'] = giga_schooldata['giga_school_upload_speed'] - giga_schooldata['ookla_upload_speed']
giga_schooldata['latency_difference'] = giga_schooldata['giga_school_latency'] - giga_schooldata['ookla_lat_ms']
```

If the value is positive in the difference, school is performing well whereas if the difference value is negative, there can be rooms for improvement
